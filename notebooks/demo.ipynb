# TS Attractor NumPy demo

Sliders require **ipywidgets**. Install: `pip install ipywidgets`.

If widgets are unavailable, use the plain Python calls in the second cell.

```python
import sys
from pathlib import Path
ROOT = Path("..").resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import numpy as np
from ts_attractor import numpy_demo as nd

try:
    import ipywidgets as w
    from IPython.display import display

    dim = w.IntSlider(value=128, min=32, max=512, step=32, description="dim")
    steps = w.IntSlider(value=24, min=10, max=40, step=1, description="Euler steps")
    tau = w.FloatSlider(value=1.0, min=0.2, max=3.0, step=0.1, description="tau")
    prompt = w.Text(value="hello attractor", description="prompt")

    out = w.Output()

    def run(_=None):
        with out:
            out.clear_output(wait=True)
            rng = np.random.default_rng(0)
            d = int(dim.value)
            diff = nd.make_diffusion_matrix(d, rng)
            sig = nd.text_to_signal(prompt.value, d)
            s = nd.integrate(diff, sig, dim=d, num_steps=int(steps.value))
            att = rng.standard_normal((8, d))
            logits = nd.proto_logits(s, att, tau=float(tau.value))
            print("||s||:", float(np.linalg.norm(s)))
            print("logits[:5]:", logits[:5])

    for x in (dim, steps, tau, prompt):
        x.observe(run, names="value")
    display(dim, steps, tau, prompt, out)
    run()
except ImportError:
    print("ipywidgets not installed; run logic manually below.")
```
